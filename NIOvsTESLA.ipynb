```python
# Import necessary libraries
import yfinance as yf
from IPython.display import display, Markdown
```

```python
NIO = yf.Ticker("NIO")
```

```python
NIO_share= NIO.history(period="1y")
```

```python
NIO_share.head()
```

```python
NIO_share.reset_index(inplace=True)
```

```python
NIO_share.plot(x="Date", y="Open")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Plot Tesla share prices
plt.plot(Tesla_share['Date'], Tesla_share['Open'], label='Tesla', color='blue')

# Plot NIO share prices
plt.plot(NIO_share['Date'], NIO_share['Open'], label='NIO', color='red')

# Adding titles and labels
plt.title('Tesla and NIO Share Prices')
plt.xlabel('Date')
plt.ylabel('Open Price')

# Adding a legend
plt.legend()

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
